```python
import pandas as pd
import numpy as np
from datetime import datetime
from urllib.request import urlopen
import json
```

```python
pm19 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=81102&bdate=20190101&edate=20191231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
pm20 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=81102&bdate=20200101&edate=20201231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
pm21 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=81102&bdate=20210101&edate=20211231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')

pm19 = json.loads(pm19.read())
pm20 = json.loads(pm20.read())
pm21 = json.loads(pm21.read())

pmdata = pd.DataFrame(pm19['Data']+pm20['Data']+pm21['Data'])
```

```python
oz19 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=44201&bdate=20190101&edate=20191231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
oz20 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=44201&bdate=20200101&edate=20201231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
oz21 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=44201&bdate=20210101&edate=20211231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')

oz19 = json.loads(oz19.read())
oz20 = json.loads(oz20.read())
oz21 = json.loads(oz21.read())

ozdata = pd.DataFrame(oz19['Data']+oz20['Data']+oz21['Data'])
```

```python
co19 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42101&bdate=20190101&edate=20191231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
co20 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42101&bdate=20200101&edate=20201231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
co21 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42101&bdate=20210101&edate=20211231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')

co19 = json.loads(co19.read())
co20 = json.loads(co20.read())
co21 = json.loads(co21.read())

codata = pd.DataFrame(co19['Data']+co20['Data']+co21['Data'])
```

```python
so19 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42401&bdate=20190101&edate=20191231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
so20 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42401&bdate=20200101&edate=20201231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
so21 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42401&bdate=20210101&edate=20211231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')

so19 = json.loads(so19.read())
so20 = json.loads(so20.read())
so21 = json.loads(so21.read())

sodata = pd.DataFrame(so19['Data']+so20['Data']+so21['Data'])
```

```python
no19 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42602&bdate=20190101&edate=20191231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
no20 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42602&bdate=20200101&edate=20201231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')
no21 = urlopen('https://aqs.epa.gov/data/api/dailyData/byBox?email=thomas.e.abraham@gmail.com&key=bayram32&param=42602&bdate=20210101&edate=20211231&minlat=41.59&maxlat=42.14&minlon=-87.93&maxlon=-87.52')

no19 = json.loads(no19.read())
no20 = json.loads(no20.read())
no21 = json.loads(no21.read())

nodata = pd.DataFrame(no19['Data']+no20['Data']+no21['Data'])
```

```python
pmdata['site_id'] = pmdata['state_code']+pmdata['county_code']+pmdata['site_number']
ozdata['site_id'] = ozdata['state_code']+ozdata['county_code']+ozdata['site_number']
codata['site_id'] = codata['state_code']+codata['county_code']+codata['site_number']
sodata['site_id'] = sodata['state_code']+sodata['county_code']+sodata['site_number']
nodata['site_id'] = nodata['state_code']+nodata['county_code']+nodata['site_number']
```

```python
new = pd.concat([pmdata,ozdata], ignore_index=True)
newer = pd.concat([new,codata], ignore_index=True)
newest = pd.concat([newer,sodata], ignore_index=True)
year = pd.concat([newest,nodata], ignore_index=True)
```

```python
oz = ozdata[['site_id','date_local','arithmetic_mean','parameter_code']]
oz = oz.rename(index=str, columns={"arithmetic_mean":"o3",'parameter_code':'parameter'})
co = codata[['site_id','date_local','arithmetic_mean','parameter_code']]
co = co.rename(index=str, columns={"arithmetic_mean":"co",'parameter_code':'parameter'})
so = sodata[['site_id','date_local','arithmetic_mean','parameter_code']]
so = so.rename(index=str, columns={"arithmetic_mean":"so2",'parameter_code':'parameter'})
no = nodata[['site_id','date_local','arithmetic_mean','parameter_code']]
no = no.rename(index=str, columns={"arithmetic_mean":"no2",'parameter_code':'parameter'})
pm = pmdata[['site_id','date_local','arithmetic_mean','parameter_code']]
pm = pm.rename(index=str, columns={"arithmetic_mean":"pm10",'parameter_code':'parameter'})


values = oz.append(co)
values = values.append(so)
values = values.append(no)
values = values.append(pm)
values = values[['site_id','date_local','parameter','o3','co','so2','no2','pm10']]
values = values.rename(index=str, columns={"site_id":"Site", "date_local":"Date"})
values['Date'] =  pd.to_datetime(values['Date'], format='%Y-%m-%d')
values[['o3', 'co', 'so2', 'no2', 'pm10']].replace('0.0', '', inplace=True)
values['Date'].replace('', np.nan, inplace=True)
values.dropna(subset=['Date'], inplace=True)

# values['pm10'] = (values['pm10'])*(.001)
# values['no2'] = (values['no2'])*(.1)
# values['so2'] = (values['so2'])*(.1)
# values['co'] = (values['co'])*(10)
# values['o3'] = (values['o3'])*(10)

values = values.fillna('')
```

```python
values = values.groupby(['Date','Site']).agg({'parameter':'first','o3': 'first', 'co':'first' , 'so2':'first','no2':'first','pm10':'first'}).reset_index()
values = pd.DataFrame(data=values)

# values['date_local'] = values['date_local'].dt.strftime('%m/%d/%Y %H:%M')
```

```python
values.to_csv('data/values.csv')
```
