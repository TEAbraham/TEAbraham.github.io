```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
pmurl = 'data/pm10_18_20.csv'
ozurl = 'data/o3_18_20.csv'
courl = 'data/co_18_20.csv'
sourl = 'data/so2_18_20.csv'
nourl = 'data/no2_18_20.csv'

pmdata = pd.read_csv(pmurl)
ozdata = pd.read_csv(ozurl)
codata = pd.read_csv(courl)
sodata = pd.read_csv(sourl)
nodata = pd.read_csv(nourl)
```

```python
new = pd.concat([pmdata,ozdata], ignore_index=True)
newer = pd.concat([new,codata], ignore_index=True)
newest = pd.concat([newer,sodata], ignore_index=True)
year = pd.concat([newest,nodata], ignore_index=True)
```

```python
year['Site ID'].unique()
year['SITE_LATITUDE'].unique()
year['SITE_LONGITUDE'].unique()

location_coords = pd.DataFrame(data={'id':year['Site ID'].unique(), 'x':year['SITE_LATITUDE'].unique(), 'y':year['SITE_LONGITUDE'].unique()})
location_coords.set_index('id')
```

```python
location_coords=location_coords.drop(location_coords.index[3])

location_coords=location_coords.set_index('id')

location_coords.to_csv('data/location_coords.csv')
```

```python
oz = ozdata[['Site ID','Date','Daily Max 8-hour Ozone Concentration','AQS_PARAMETER_CODE']]
oz = oz.rename(index=str, columns={"Daily Max 8-hour Ozone Concentration":"o3",'AQS_PARAMETER_CODE':'parameter'})
co = codata[['Site ID','Date','Daily Max 8-hour CO Concentration','AQS_PARAMETER_CODE']]
co = co.rename(index=str, columns={"Daily Max 8-hour CO Concentration":"co",'AQS_PARAMETER_CODE':'parameter'})
so = sodata[['Site ID','Date','Daily Max 1-hour SO2 Concentration','AQS_PARAMETER_CODE']]
so = so.rename(index=str, columns={"Daily Max 1-hour SO2 Concentration":"so2",'AQS_PARAMETER_CODE':'parameter'})
no = nodata[['Site ID','Date','Daily Max 1-hour NO2 Concentration','AQS_PARAMETER_CODE']]
no = no.rename(index=str, columns={"Daily Max 1-hour NO2 Concentration":"no2",'AQS_PARAMETER_CODE':'parameter'})
pm = pmdata[['Site ID','Date','Daily Mean PM10 Concentration','AQS_PARAMETER_CODE']]
pm = pm.rename(index=str, columns={"Daily Mean PM10 Concentration":"pm10",'AQS_PARAMETER_CODE':'parameter'})
```

```python
values = oz.append(co)
values = values.append(so)
values = values.append(no)
values = values.append(pm)
values = values[['Site ID','Date','parameter','o3','co','so2','no2','pm10']]
values = values.rename(index=str, columns={"Site ID":"Site"})
values['Date'] =  pd.to_datetime(values['Date'], format='%m/%d/%Y')
values['Site'].replace('END OF FILE','',inplace=True)
values[['o3', 'co', 'so2', 'no2', 'pm10']].replace('0.0', '', inplace=True)
values['Date'].replace('', np.nan, inplace=True)
values.dropna(subset=['Date'], inplace=True)
# values['pm10'] = (values['pm10'])*(.001)
# values['no2'] = (values['no2'])*(.1)
# values['so2'] = (values['so2'])*(.1)
# values['co'] = (values['co'])*(10)
# values['o3'] = (values['o3'])*(10)
values = values.fillna('')
```

```python
values = values.groupby(['Date','Site']).agg({'parameter':'first','o3': 'first', 'co':'first' , 'so2':'first','no2':'first','pm10':'first'}).reset_index()
values = pd.DataFrame(data=values)

# values['Date'] = values['Date'].dt.strftime('%m/%d/%Y %H:%M')
```

```python
values.to_csv('data/values.csv')
```
